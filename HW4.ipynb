```python
#read in the data file; make labeled arrays of each column
test = ascii.read('gravDataFull.dat')

i = 5
percent = []
while i <= 122:
    distcol = 'col'+str(i)
    maxcol = 'col'+str(i+1)
    mincol = 'col'+str(i+2)
    percent.append((max(test[maxcol])-min(test[mincol]))/test[distcol][0]*100)
    i = i+3
```

```python
x = np.linspace(3.0, 3.6, 40)
plt.scatter(x, percent)
#plt.ylim(1.16, 1.2)
plt.ylabel('Smear Percentage')
plt.xlabel('Orbital Radius (AU)')
plt.savefig('../school_stuff/CompPhysFigures/KirkwoodGap3-3.png')
```
